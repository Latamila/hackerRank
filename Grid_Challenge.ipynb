Dada uma grade quadrada de caracteres no intervalo ascii[az], reorganize os elementos de cada linha em ordem alfabética, crescente. Determine se as colunas também estão em ordem alfabética crescente, de cima para baixo. Retorne YES se estão ou NO, se não.

image.png

As linhas já estão em ordem alfabética. As colunas:

a a e,

b d f

c e g

também estão em ordem alfabética, então a resposta seria YES.

Somente elementos dentro da mesma linha podem ser reorganizados.

Eles não podem ser movidos para uma linha diferente.

gridChallenge possui os seguintes parâmetros:

string grid[n]: uma matriz de strings

Devoluções

string: ou YES ou NO

Cada string consiste em letras minúsculas no intervalo ascii[az]

Para cada caso de teste, imprima em uma linha separada YESse é possível reorganizar a grade em ordem alfabética crescente em linhas e colunas, ou NO não.

```python
import math
import os
import random
import re
import sys
```

```python
def is_column_sorted(column):
    return all(column[i] <= column[i + 1] for i in range(len(column) - 1))

def gridChallenge(t, grids):
    results = []
    for i in range(t):
        sorted_grid = [sorted(row) for row in grids[i]]
        is_sorted = all(is_column_sorted(column) for column in zip(*sorted_grid))
        results.append('YES' if is_sorted else 'NO')
    return results

if __name__ == '__main__':

    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    t = int(input().strip())

    grids = []
    for _ in range(t):
        n = int(input().strip())
        grid = [input().strip() for _ in range(n)]
        grids.append(grid)

    results = gridChallenge(t, grids)

    fptr.write('\n'.join(results) + '\n')

    fptr.close()
```

#1. Função is_column_sorted(column)
Esta função verifica se uma coluna (ou lista) está ordenada em ordem não decrescente. Ela itera sobre os elementos da coluna e compara cada elemento com o próximo. Se todos os elementos forem menores ou iguais ao próximo elemento, a função retorna True, indicando que a coluna está ordenada.

image.png

all(): Essa função retorna True se todos os elementos no iterável (no caso, a expressão geradora) forem verdadeiros. No código, ela verifica se todas as comparações column[i] <= column[i + 1] são verdadeiras.

for i in range(len(column) - 1): Isso itera sobre os índices da coluna, e column[i] <= column[i + 1] compara cada elemento com o próximo.

#2. Função gridChallenge(t, grids)

Esta função recebe o número de casos de teste t e uma lista de grids "grids". Cada grid é uma lista de strings.
