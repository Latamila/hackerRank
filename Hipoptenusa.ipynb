```python
n = int(input('digite o numero positivo: '))
h_quadrado = 0
numeros_inteiros = set()

for i in range(1,n+1):
  for j in range(1, n+1):
    h_quadrado = math.sqrt((i**2)+(j**2))
    print(f'a hipotenusa dos catetos {i} e {j} é {h_quadrado}')
    if h_quadrado.is_integer():
        print(f'{h_quadrado} é um número inteiro.')
        # Armazena o número inteiro no conjunto (set)
        if h_quadrado <= n:
          numeros_inteiros.add(int(h_quadrado))
    else:
        print(f'{h_quadrado} é um número decimal.')
print(numeros_inteiros)

```

```python
def e_hipotenusa(n):
  h_quadrado = 0
  numeros_inteiros = set()

  for i in range(1,n+1):
    for j in range(1, n+1):
      h_quadrado = math.sqrt((i**2)+(j**2))
      if h_quadrado.is_integer():
          if h_quadrado <= n:
            numeros_inteiros.add(int(h_quadrado))

  return numeros_inteiros

```

```python
e_hipotenusa(25)
```

```python
def soma_hipotenusas(n):
  conjunto = e_hipotenusa(n)
  lista = list(conjunto)
  soma = 0
  for i in range(len(lista)):
    soma = soma + lista[i]
  return soma

```

```python
soma_hipotenusas(25)
```
