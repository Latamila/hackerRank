Sean inventou um jogo envolvendo uma matriz 2n * 2n onde cada célula da matriz contém um número inteiro. Ele pode reverter qualquer uma de suas linhas ou colunas quantas vezes quiser. O objetivo do jogo é maximizar a soma dos elementos da submatriz n *n localizada no quadrante superior esquerdo da matriz.

Dadas as configurações iniciais para matrizes q, ajude Sean a reverter as linhas e colunas de cada matriz da melhor maneira possível para que a soma dos elementos no quadrante superior esquerdo da matriz seja máxima.


```python
matrix = [[112, 42, 83, 119], [56, 125, 56, 49], [15, 78, 101, 43], [62, 98, 114, 108]]
```

```python
sum = 0

n = len(matrix) #recebe o comprimento da matrix

for i in range(n // 2): #pega até a metade das linhas
    for j in range(n // 2): #pega até a metade das colunas
        sum += max(matrix[i][j],
                   max(matrix[i][n-1-j],
                       max(matrix[n-1-i][j],
                           matrix[n-1-i][n-1-j])))
        print(sum)
```

Temos que retornar a soma de todos os numeros max que possuem as mesmas cores, como primeiro temos que encontrar max de [112,119,62,108] e assim por diante.

Estamos percorrendo apenas metade da lista porque temos que encontrar a soma do canto superior da matriz.

Ao percorrer uma determinada lista de lista, obtemos a seguinte resposta na variavel sum.


```python
def flippingMatrix(matrix):
    sum = 0 #vai acumular o valor da soma dos maiores numeros
    n= len(matrix)
    for i in range(n // 2):
        for j in range(n // 2):
            sum += max(matrix[i][j], max(matrix[i][n-1-j], max(matrix[n-1-i][j], matrix[n-1-i][n-1-j])))

    return sum
```

```python
flippingMatrix(matrix)
```
